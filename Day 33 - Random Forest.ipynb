```python
import pandas as pd
import numpy as np 
pd.set_option('display.max_columns',None)
pd.set_option('display.max_row',None)
```

```python
ctg = pd.read_csv(r'CTG.csv')
```

```python
ctg.head()
```

```python
ctg.isnull().sum()
```

```python
ctg.shape
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
ctg[ctg.select_dtypes(include=['object']).columns] = ctg[ctg.select_dtypes(include=['object']).columns].apply(le.fit_transform)
```

```python
ctg.info()
```

```python
# SAmpling
```

```python
from sklearn.model_selection import train_test_split
ctg_train, ctg_test = train_test_split(ctg, test_size = .2)
```

```python
ctg_train_x = ctg_train.iloc[:,0:-1]
ctg_train_y = ctg_train.iloc[:,-1]

ctg_test_x = ctg_test.iloc[:,0:-1]
ctg_test_y = ctg_test.iloc[:,-1]
```

```python
# Decision Tree
```

```python
from sklearn.tree import DecisionTreeClassifier
dtree = DecisionTreeClassifier()
dtree.fit(ctg_train_x,ctg_train_y)
```

```python
dec_pred = dtree.predict(ctg_test_x)
dec_pred
```

```python
#Confusion Matrix
```

```python
from sklearn.metrics import confusion_matrix
```

```python
dtree_tab = confusion_matrix(ctg_test_y, dec_pred)
dtree_tab
```

```python
# Accuracy
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(ctg_test_y , dec_pred)
```

```python
# random Forest
```

```python
from sklearn.ensemble import RandomForestClassifier
ran_for = RandomForestClassifier(n_estimators=50,criterion='gini')
ran_for.fit(ctg_train_x , ctg_train_y)
```

```python
pred_ran = ran_for.predict(ctg_test_x)
pred_ran
```

```python
tab_ran = confusion_matrix(ctg_test_y , pred_ran)
tab_ran
```

```python
accuracy_score(ctg_test_y , pred_ran)
```
