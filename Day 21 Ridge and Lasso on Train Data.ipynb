```python
import pandas as pd 
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
import numpy as np
```

```python
tt = pd.read_csv(r"train.csv")
```

```python
tt.head()
```

```python
tt = tt.drop(['ID'], axis=1)
```

```python
tt.isnull().sum()
```

```python
tt.shape
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
tt[tt.select_dtypes(include=['object']).columns] = tt[tt.select_dtypes(include=['object']).columns].apply(le.fit_transform)
```

```python
tt.head()
```

```python
tt.info()
```

```python
from sklearn.model_selection import train_test_split
```

```python
tt_train, tt_test = train_test_split(tt, test_size=.2)
```

```python
tt_train_x = tt_train.iloc[:, 1::]
tt_train_y = tt_train.iloc[:, 0]
```

```python
tt_test_x = tt_test.iloc[:, 1::]
tt_test_y = tt_test.iloc[:, 0]
```

```python
from sklearn.linear_model import LinearRegression
```

```python
linereg = LinearRegression()
```

```python
linereg.fit(tt_train_x, tt_train_y)
```

```python
Rsquare = linereg.score(tt_train_x, tt_train_y)
Rsquare
```

```python
N = tt_train_x.shape[0]
K = tt_train_x.shape[1]
```

```python
adj_Rsquare = 1-(1-Rsquare) * (N-1)/ (N-K-1)
adj_Rsquare
```

```python
linereg.coef_
```

```python
linereg.intercept_
```

```python
pred_train = linereg.predict(tt_train_x)
pred_test = linereg.predict(tt_test_x)
```

```python
err_test = tt_test_y - pred_test
err_train = tt_train_y - pred_train
```

```python
mse_test = np.mean(np.square(err_test))
mse_test
```

```python
mse_train = np.mean(np.square(err_train))
mse_train
```

```python
rmse_test = np.sqrt(mse_test)
rmse_test
```

```python
rmse_train = np.sqrt(mse_train)
rmse_train
```

```python
mape_test = np.mean(np.abs(err_test * 100 / tt_test_y))
mape_test
```

```python
mape_train = np.mean(np.abs(err_train * 100 / tt_train_y))
mape_train
```
