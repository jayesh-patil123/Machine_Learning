```python
from sklearn.cluster import KMeans
```

```python
kmeans_sns = KMeans(n_clusters=5)
```

```python
kmeans_sns.fit(sns)
```

```python
# we will build the plot.
# run a loop from k=1 till k=10
```

```python
kmeans_sns.score(sns)

# While building the plot we need to ignore the minus sign i.e we are going to takjre the mod.
```

```python
sn=range(1,11); print(sn)
kmeans_sns = [KMeans(n_clusters = i) for i in sn ]
kmeans_sns
score = [kmeans_sns[i].fit(sns).score(sns) * -1 for i in range(len(kmeans_sns))]
print(score)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(sn , score , marker='*')
plt.grid()
```
