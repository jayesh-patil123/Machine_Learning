```python
del mk['CustomerID']
```

```python
mk.Genre.replace({'Female' : 1 , 'Male' : 0}, inplace=True)
```

```python
mk.head()
```

```python
import scipy
```

```python
from scipy.cluster.hierarchy import dendrogram, linkage
```

```python
z1 = linkage(mk,'ward' )
```

```python
z1.shape
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize = (10, 10))
dendrogram(z1 , p=16, truncate_mode='lastp')
plt.title('Hirarchical Clustering')
plt.xlabel('Cluster Size')
plt.ylabel('Distance')
```

```python
# Agglomerative Clustering
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
clusters = AgglomerativeClustering(n_clusters=5)

# Heere default value is 2
```

```python
clusters.fit_predict(mk)
```

```python
# Lets plot them and see how clusters are distrubuted
```

```python
plt.scatter(mk['Age'], mk ['Spending Score (1-100)'], c = clusters.labels_ , cmap='rainbow')
```

```python
fig, ax= plt.subplots()
sc = ax.scatter(mk['Annual Income (k$)'] , mk['Spending Score (1-100)'],  c = clusters.labels_ , cmap='rainbow')
ax.legend(*sc.legend_elements(), title ='clusters')
```
