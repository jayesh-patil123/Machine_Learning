Note:- There one problem in the linear regression is that there is high chance that model will getting Overfitting.

- Overfitting is when the model will give very good results on train data and bad results on test data.

- In overfitting the model will try to memik the data. (:. mean it will adjust the all data point cover all the datapoints)

-Lets say there are two points in my training data and iam having two features or variables X & Y..

            Train Accuracy :- 90-98 %
            Test Accuracy :- 66-65 %
            
            Low Bias
            High Variance
            
* Underfitting :- 
            Train Accuracy :- 60%
            test Accuracy :- 62%
            
            High Bias

            Low Bias

* Reason for Overfitting

1] Data used for Training is not cleaned properly and it contain noise in it.

2] The model has a high variance there will be chance to get overfittiong in it.

3] The size of training data used is very small.

4] The model is too complex.
