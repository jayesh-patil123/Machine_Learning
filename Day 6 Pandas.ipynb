```python
import numpy as np
import pandas as pd
```

```python
# Merging
```

```python
left = pd.DataFrame({'key': ['K0', 'K8', 'K2', 'K3'],
                    'A': ['A1','A2','A3','A4'],
                    'B': ['B1','B2','B3','B4']})
left

right = pd.DataFrame({'key': ['K0', 'K8', 'K2', 'K3'],
                    'C': ['C1','C2','C3','C4'],
                    'D': ['D1','D2','D3','D4']})
right
```

```python
left
```

```python
right
```

```python
inner = pd.merge(left,right,how='inner',on='key')
inner
```

```python
leftjoin = pd.merge(left,right,how='left',on='key')
leftjoin
```

```python
rightjoin = pd.merge(left,right,how='right',on='key')
rightjoin
```

```python
# Groupby()
```

```python
data = {'Company':['GOOG', 'GOOG','MSFT', 'MSFT', 'FB', 'FB'],
       'Person': ['SAM', 'CHARLIE', 'AMY','VENSSA','CARL','SARAH'],
       'SALES': [200,120,340,124,243,350]}
df = pd.DataFrame(data)
df
```

```python
bycomp = df.groupby('Company')
bycomp
```

```python
bycomp.mean()
```

```python
bycomp.median()
```

```python
bycomp.std()
```

```python
age_list = [['Afghanistan', 1952, 8425333, 'Asia'],
            ['Australia', 1957, 9712569, 'Oceania'],
            ['Brazil', 1962, 76039390, 'mericas'],
            ['China', 1957, 637408000, 'Asia'],
            ['France', 1957, 44310863, 'Europe'],
            ['India', 1952, 3.72e+08, 'Asia'],
            ['United States', 1957, 171984000, 'Americas']]

# creating a pandas dataframe

df = pd.DataFrame(age_list, columns=['Country', 'Year', 'Population', 'Continent'])
df
```

```python
# Sorting by column
```

```python
df.sort_values(by=['Country'])
```

```python
df.sort_values(by=['Population'], ascending=False)
```
