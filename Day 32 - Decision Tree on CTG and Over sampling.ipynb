```python
W = pd.read_csv(r'wine.csv')
```

```python
from sklearn.model_selection import train_test_split
W_train, W_test = train_test_split(W, test_size=25)
```

```python
W_train_x = W_train.iloc[:,1::]
W_train_y = W_train.iloc[:,-1]

W_test_x = W_test.iloc[:,1::]
W_test_y = W_test.iloc[:,-1]
```

```python
W_train_x.head()
```

```python
W_train_y.head()
```

```python
W_test_x.head()
```

```python
W_test_y.head()
```

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier(criterion='gini',max_depth=None, min_samples_split=100 )
dt.fit(W_train_x, W_train_y)
```

```python
dec_pred = dt.predict(W_test_x)
dec_pred
```

```python
from sklearn.metrics import confusion_matrix
```

```python
dt_tab = confusion_matrix(W_test_y , dec_pred)
dt_tab
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(W_test_y, dec_pred)*100
```
