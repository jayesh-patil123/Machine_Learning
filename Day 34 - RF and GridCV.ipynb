```python
import pandas as pd
import numpy as np 
pd.set_option('display.max_columns',None)
pd.set_option('display.max_row',None)
```

```python
B = pd.read_csv(r'Patient_with_abnormal_bloodpressure.csv')
```

```python
B.head()
```

```python
B.isnull().sum()
```

```python
B.shape
```

```python
B = B.drop(['Genetic_Pedigree_Coefficient'], axis=1)
B = B.drop(['Pregnancy'], axis=1)
B = B.drop(['alcohol_consumption_per_day'], axis=1)
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
B[B.select_dtypes(include=['object']).columns] = B[B.select_dtypes(include=['object']).columns].apply(le.fit_transform)
```

```python
B.info()
```

```python
# SAmpling
```

```python
from sklearn.model_selection import train_test_split
B_train, B_test = train_test_split(B, test_size = .2)
```

```python
B_train_x = B_train.iloc[:,0:-1]
B_train_y = B_train.iloc[:,-1]

B_test_x = B_test.iloc[:,0:-1]
B_test_y = B_test.iloc[:,-1]
```

```python
# Decision Tree
```

```python
from sklearn.tree import DecisionTreeClassifier
dtree = DecisionTreeClassifier()
dtree.fit(B_train_x,B_train_y)
```

```python
dec_pred = dtree.predict(B_test_x)
dec_pred
```

```python
#Confusion Matrix
```

```python
from sklearn.metrics import confusion_matrix
```

```python
dtree_tab = confusion_matrix(B_test_y, dec_pred)
dtree_tab
```

```python
# Accuracy
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(B_test_y , dec_pred)
```

```python
# random Forest
```

```python
from sklearn.ensemble import RandomForestClassifier
ran_for = RandomForestClassifier()
ran_for.fit(B_train_x , B_train_y)
```

```python
pred_ran = ran_for.predict(B_test_x)
pred_ran
```

```python
tab_ran = confusion_matrix(B_test_y , pred_ran)
tab_ran
```

```python
accuracy_score(B_test_y , pred_ran)
```
